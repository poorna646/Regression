```python
import numpy as np
import matplotlib.pyplot as plt
```

REGULARIZATION TECHNIQUE

RIDGE REGRESSION

*Bringing data into particular range, add a penalty term with loss/error to regularize.
*If penalty term is squared it is ridge 
*It doesn't reach 0 easily bcoz of sqaured value 
*Has squared coeff

LASSO REGRESSION

*Can reach 0 bcoz of modulus
*It is best bcoz error can be brought easily to 0 
*Has modulus coeff 

```python
x=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
x_b=np.c_[np.ones((x.shape[0],1)),x]#column stack-c_
x_b
```

```python
alpha_ridge=1.0
alpha_lasso=0.1
num_iterations=1000
learning_rate=0.001
```

```python
def lasso_regression(x,y,alpha, num_iterations,learning_rate): #alpha-penalty value
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred= x.dot(theta)
    gradient = (x.T.dot(y_pred - y)+ alpha * np.sign(theta))
    theta -= learning_rate * gradient
  return theta
```

```python
def ridge_regression(x,y,alpha, num_iterations,learning_rate):
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred= x.dot(theta)
    gradient = (x.T.dot(y_pred - y)+ alpha * theta)/m
    theta -= learning_rate * gradient
  return theta
```

```python
theta_ridge = ridge_regression(x_b,y, alpha_ridge,num_iterations, learning_rate)
theta_lasso = lasso_regression(x_b,y,alpha_ridge, num_iterations, learning_rate)
```

```python
plt.scatter(x,y,label='Data')
plt.plot(x,x_b.dot(theta_ridge), label=f'Ridge (L2) Regression',color='red')
plt.plot(x,x_b.dot(theta_lasso), label=f'Lasso (L1) Regression',color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
