```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

POLYNOMIAL REGRESSION
polynomial- deals with non-linear datas(line can't fit- so transform by giving degree)


```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data
```

```python
x=data.iloc[:,1:2].values
x
```

```python
y=data.iloc[:,2:].values
y
```

```python
#way to import linear regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures #rise the degree
poly_reg2=PolynomialFeatures(degree=2)
x_poly=poly_reg2.fit_transform(x) #we can't use y  bcoz x power only changes in general formula y=b0+b1x1+b2x1^2+...
#changing polynomial into linear
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures #rise the degree
poly_reg3=PolynomialFeatures(degree=3)
x_poly3=poly_reg3.fit_transform(x) #we can't use y  bcoz x power only changes in general formula y=b0+b1x1+b2x1^2+...
#changing polynomial into linear
lin_reg_3=LinearRegression()
lin_reg_3.fit(x_poly3,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regreesion')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(x_poly),color='blue')
plt.plot(x,lin_reg_3.predict(x_poly3),color='green')
plt.title('Linear Regreesion')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
x=data1.drop('Y house price of unit area',axis=1)
x
```

```python
y=data1['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2, include_bias=False)
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=101) #30% assigned to test
```

```python
polymodel=LinearRegression()#1. model created
```

```python
#fit the model
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'y_test': y_test,'y_pred':y_pred,'Residuals':(y_test  - y_pred)})
```

```python
#metrics - provides quality of prediction
Poly_MSE = metrics.mean_absolute_error(y_test,y_pred)
poly_MSE= metrics.mean_squared_error(y_test,y_pred)
Poly_MSE =np.sqrt(Poly_MSE) #relative squared error
Poly_MSE
```
