```python
data['age'].replace(np.nan,data['age'].astype('float').mean(axis=0),inplace=True)
data['body'].replace(np.nan,data['body'].astype('float').mean(axis=0),inplace=True)
data['home.dest'].fillna('Unknown', inplace=True)
data['boat'].fillna('0', inplace=True)
```

```python
data["sex"] = data["sex"].replace(to_replace=data["sex"].unique(),value=[1,0,2])
```

```python
data= pd.get_dummies(data,drop_first=True)
```

```python
y_train = data['survived']
x_train = data[['pclass', 'sex','age','sibsp','parch','fare']]
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_target, test_target =train_test_split(x_train,y_train,train_size=0.8)  # 80% datas for training
```

```python
from sklearn.model_selection import train_test_split

X = data.loc[:, ['pclass', 'age', 'sex']]
y = data.survived

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
X_test = X_test.dropna()
y_test = y_test.loc[X_test.index]
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("accuracy score: ",accuracy_score(y_test,pred))
```
