```python
#factorize for all columns , _ is 2 avoid another varaible
data['buying'],_ = pd.factorize(data['buying'])
data['persons'],_ = pd.factorize(data['persons'])
data['maint'],_ = pd.factorize(data['maint'])
data['doors'],_ = pd.factorize(data['doors'])
data['lug_boot'],_ = pd.factorize(data['lug_boot'])
data['safety'],_ = pd.factorize(data['safety'])

data.head()
```

```python
#encoding-label,one-hot
x=data.iloc[:,0:-1]
y=data.iloc[:,-1]
x
y
```

```python
data.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.3,random_state=0) #random_state- doesn't mix test and train datas,default value pf random_state is 42
```

```python
x.columns
```

```python
#define the model
from sklearn import tree,metrics
dtree = tree.DecisionTreeClassifier(criterion='gini') # entropy or gini can be given for finding best split
dtree.fit(x_train , y_train)
```

```python
y_train_pred = dtree.predict(x_train)
```

```python
y_test_pred = dtree.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_train , y_train_pred)
print('Train accuracy: {:.2f}'.format(accuracy))
```

```python
accuracy = metrics.accuracy_score(y_test , y_test_pred)
print('Test accuracy: {:.2f}'.format(accuracy))
```
